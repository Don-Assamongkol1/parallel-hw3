```python
import matplotlib.pyplot as plt 
```

# Idle Overhead Test

```python
lock_to_time_ratio = [
("TAS lock",6.276697583852443),
("Pthread lock",12.206150612493127),
("Anderson lock",6.256511288926657),
("TTAS lock",6.145553889553269)
]


for (lock, time) in lock_to_time_ratio:
    plt.bar(lock,
    time,
    # marker='x'
    )

plt.title("Idle Overhead Test")
plt.ylabel("Ratio of execution time with lock to serial execution time")
```

# Lock Scaling Test

```python
Ns = [1, 2, 4, 8, 14]

lock_to_times = [
    ("TAS lock", [10.885361595766273,
    101.17283704811804,
    207.37497427589892,
    402.92406276169504,
    606.8405785366881]),

    ("Pthread Lock", [11.725724710021275,
    70.87606224300332,
    67.82721608395288,
    82.065709968653,
    85.04004503597014]),

    ("Anderson Lock", [6.229276673156577,
    179.1094299427936,
    171.87755102347646,
    174.52193459974293,
    203.0665722973532]),

    ("TTAS lock", [11.134732046877703,
    87.22191823464411,
    139.41508093093884,
    237.4805578717103,
    450.8879144910432])
]

for (lock, time) in lock_to_times:
    plt.plot(Ns,
    time,
    marker='x'
    )

plt.legend(["TAS", "Pthread", "Anderson", "TTAS"])
plt.title("Lock Scaling Test")
plt.xlabel("Number of Threads (n)")
plt.ylabel("Ratio of execution time with lock to serial execution time")
```
